```python
import pandas as pd
import statsapi
from pprint import pprint
import requests
from bs4 import BeautifulSoup
```

```python
espn_url = "https://www.espn.com/mlb/war/leaders/_/type/seasonal/year/20"
```

```python
year_list = ['19', '18', '17', '16', '15', '14', '13', '12', '11', '10']
df_list = []
```

```python
for year in year_list:
    query_url = espn_url+year
    qdfl = pd.read_html(query_url)
    qdfl[0]['season'] = year
    df_list.append(qdfl[0])
```

```python
data = pd.concat(df_list)
data.head()
```

```python
cols = ['rank', 'player', 'WAR', 'OFF', 'DEF', 'WAA', 'TRPG', 'ORPG', 'RAA', 'WAAWP', 'Season']
```

```python
data.columns = cols
del data['rank']
data.head()
```

```python
data = data.dropna(subset=['player'])
```

```python
data = data[~data.player.str.contains("WAR")]
```

```python
data.head()
```

```python
# new data frame with split value columns 
new = data["player"].str.split(" ", n = 1, expand = True) 
  
# making separate first name column from new data frame 
data["First Name"]= new[0] 
data["Last Name"]= new[1] 
  
data = data.drop(data.index[0])
data.reset_index()
data.head()
```

```python
data.drop_duplicates(subset ="player", keep = False, inplace = True)
```

```python
data.to_csv('player_list.csv')
```
